#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one.
```python
# For the date early in the dataset, measure the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one

bitstamp_early = bitstamp_df["Close"].loc['2018-01-01' : '2018-01-01']
coinbase_early = coinbase_df['Close'].loc['2018-01-01' : '2018-01-01']

arbitrage_spread_early =  coinbase_early - bitstamp_early 

# # Use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame
# YOUR CODE HERE

arbitrage_spread_early[arbitrage_spread_early > 0].describe()
```

```python
bitstamp_middle = bitstamp_df["Close"].loc['2018-02-28' : '2018-02-28']
coinbase_middle = coinbase_df['Close'].loc['2018-02-28' : '2018-02-28']

arbitrage_spread_middle = bitstamp_middle - coinbase_middle

arbitrage_spread_middle[arbitrage_spread_middle > 0].describe()
```

```python
bitstamp_late = bitstamp_df["Close"].loc['2018-03-28' : '2018-03-28']
coinbase_late = coinbase_df['Close'].loc['2018-03-28' : '2018-03-28']

arbitrage_spread_late = bitstamp_late - coinbase_late

arbitrage_spread_late[arbitrage_spread_late > 0].describe()
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).

spread_return_early= arbitrage_spread_early[arbitrage_spread_early > 0] / bitstamp_early


# YOUR CODE HERE

# Review the spread return DataFrame
# YOUR CODE HERE

spread_return_early.describe()
```

```python
spread_return_middle = arbitrage_spread_middle[arbitrage_spread_middle > 0] / coinbase_middle
# I use coinbase in the denominator for this particular date because coinbase was the less costly exhange for Bitcoin.

spread_return_middle.describe()
```

```python
spread_return_late = arbitrage_spread_late[arbitrage_spread_late > 0] / coinbase_late

spread_return_late.describe()
```

#### 3. For each of the three dates, narrow down your trading opportunities even further. To do so, determine the number of times your trades with positive returns exceed the 1% minimum threshold that you need to cover your costs.
